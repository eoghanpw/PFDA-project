```python
import csv
import requests
from io import StringIO
import re
import pandas as pd

# File containing csv urls.
txt_file = "./download-file-list-daily.txt"

# Dictionary to store the dataframes.
dataframes = {}

# Regex pattern to find start of useful data in csv files.
pattern = "^date(?!:)"

# Columns to import.
columns = ["date", "wdsp"]

# Read the urls from the txt file and strip new lines.
with open(txt_file) as f:
    csv_urls = [line.strip() for line in f if line.strip()]

# Loop through the urls.
for url in csv_urls:
    # Send get request to url.
    response = requests.get(url)
    
    # Check if request successful.
    if response.status_code == 200:
        # Treat response content as file like object.
        csv_file = StringIO(response.text)

        # CSV reader.
        csv_reader = csv.reader(csv_file)
        
        # Search each file for start of useful data.
        for row_number, row in enumerate(csv_reader, start=0):
            row_string = ",".join(row)
            if any(re.search(pattern, row_string) for row_string in row):
                print(f"Skipped {row_number} rows in {url}")
                break
        else:
            print(f"Error finding regex pattern: \"{pattern}\" in {url}")
        
    else:
        print(f"Failed to retrieve the CSV file. Status code: {response.status_code}")

    # Send get request to url.
    response = requests.get(url)
    
    # Check if request successful.
    if response.status_code == 200:
        # Treat response content as file like object.
        csv_file = StringIO(response.text)

        # CSV reader.
        csv_reader = csv.reader(csv_file)
        try:
            df_name = url.split("webdata/")[1].strip(".csv")
            dataframes[df_name] = pd.read_csv(csv_file, skiprows=row_number, usecols=columns)
            print(f"Imported {url} as {df_name}")
        except ValueError as e:
            print(f"Error reading {url}: {e}")
    else:
        print(f"Failed to retrieve the CSV file. Status code: {response.status_code}")
```
